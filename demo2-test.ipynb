```python
import datetime

# Define a base date (could be any date; just keep consistent for Day 1 vs Day 2)
base_date = datetime.date(2025, 1, 1)

def make_datetime(day_number, time_str):
    """
    day_number: 1 or 2
    time_str: string in 'HH:MM' format
    """
    # Convert day_number to a real date by adding (day_number - 1) days to the base_date
    actual_date = base_date + datetime.timedelta(days=day_number - 1)
    # Convert the time_str (HH:MM) to a time object
    time_obj = datetime.datetime.strptime(time_str, "%H:%M").time()
    # Combine date and time into a datetime
    return datetime.datetime.combine(actual_date, time_obj)

# Create datetime columns for Start and End
df_acti['Start_dt'] = df_acti.apply(
    lambda row: make_datetime(row['Day'], row['Start']), axis=1
)
df_acti['End_dt'] = df_acti.apply(
    lambda row: make_datetime(row['Day'], row['End']), axis=1
)

# Now you can drop (or keep) the original columns if you want
# df_activity.drop(['Day','Start','End'], axis=1, inplace=True)
```

```python
def make_datetime_seconds(day_number, time_str):
    """
    day_number: 1 or 2
    time_str: string in 'HH:MM:SS' format
    """
    actual_date = base_date + datetime.timedelta(days=day_number - 1)
    time_obj = datetime.datetime.strptime(time_str, "%H:%M:%S").time()
    return datetime.datetime.combine(actual_date, time_obj)

df_actig['Datetime'] = df_actig.apply(
    lambda row: make_datetime_seconds(row['day'], row['time']), axis=1
)
```
